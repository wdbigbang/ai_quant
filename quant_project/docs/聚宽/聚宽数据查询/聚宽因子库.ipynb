# 使用方法
**声明：**

**1. 在单因子分析中可以直接获取因子库中的数据；**

**2. 同时也可以通过API的形式，在其他模块中获取这些因子；**

**3. 为保证数据的连续性，所有数据基于后复权计算。**

**4. 涉及到财务数据的因子，使用季报数据进行计算**

**5. 为了防止单次返回数据时间过长，每次调用 api 请求的因子值不能超过 200000 个**

**6. 频率为天，每天05：00更新前一天数据**

**7. 提供股票的因子数据，不支持期货、指数等**

## 获取因子值
获取因子值:

#导入函数库（研究环境）
from jqfactor import get_factor_values

#导入函数库（jqdatasdk）
from jqdatasdk import get_factor_values

#取值函数
get_factor_values(securities, factors, start_date, end_date, count)

参数

securities:股票池，单只股票（字符串）或一个股票列表

factors: 因子名称，单个因子（字符串）或一个因子列表

start_date:开始日期，字符串或 datetime 对象，与 coun t参数二选一

end_date: 结束日期， 字符串或 datetime 对象，可以与 start_date 或 count 配合使用

count: 截止 end_date 之前交易日的数量（含 end_date 当日），与 start_date 参数二选一

返回

一个 dict： key 是因子名称， value 是 pandas.dataframe。

dataframe 的 index 是日期， column 是股票代码， value 是因子值

示例

```python
# 导入函数库
# from jqfactor import get_factor_values
```

```python
# 获取因子Skewness60(个股收益的60日偏度)从 2017-01-01 至 2017-03-04 的因子值
factor_data = get_factor_values(securities=['000001.XSHE'], factors=['Skewness60','DEGM','quick_ratio'], start_date='2017-01-01', end_date='2017-03-04')
```

| non_linear_size     | 非线性市值 | 描述了无法由规模因子解释的但与规模有关的收益差异，通常代表中盘股 |
| book_to_price_ratio | 账面市值比 | 描述了股票估值高低不同而产生的收益差异, 即价值因子           |
| liquidity           | 流动性     | 解释了由股票相对的交易活跃度不同而产生的收益率差异           |
| earnings_yield      | 盈利能力   | 描述了由盈利收益导致的收益差异                               |
| growth              | 成长       | 描述了对销售或盈利增长预期不同而产生的收益差异               |
| leverage            | 杠杆       | 描述了高杠杆股票与低杠杆股票之间的收益差异                   |

**因子计算方法如下所示：**

| 因子                           | 定义                                                         |
| ------------------------------ | ------------------------------------------------------------ |
| size 市值                      | 定义：公司的总市值的自然对数                                 |
| beta 贝塔                      | 定义：时间序列上的个股超额收益与股票池总收益的回归系数，被解释变量是个股的无风险收益，解释变量是市值加权的股票池总收益。 |
| momentum 动量                  | 定义：计算过去504日的超额对数收益的和，以21天为间隔进行计算。 |
| residual_volatility 残差波动率 | 定义：0.74 · DASTD + 0.16 · CMRA + 0.10 · HSIGMA 其中 DASTD: 日标准差，过去252日的超额收益的波动率，以42日为一个半衰期。 CMRA: 累积波动范围，过去12个月中个股波动率的最大值和最小值之间的差异。 HSIGMA: sigma历史波动，计算BETA时候的回归残差项的标准差。 |
| non_linear_size 非线性市值     | 定义：首先对标准化后的市值因子暴露值求立方，将得到的结果与市值因子进行加权回归的正交化处理。最后在进行缩尾和标准化。 |
| book_to_price_ratio 账面市值比 | 定义：最新一季财报的账面价值与当前市值的比值。               |
| liquidity 流动性               | 定义： 0.35 · STOM + 0.35 · STOQ + 0.3 · STOA 其中 STOM: 股票一个月换手率，过去21日的股票换手率之和的对数。 STOQ: 过去3个月平均换手率，计算过去3个月的平均STOM,并取对数。 STOA: 过去12个月平均换手率，计算过去12个月的平均STOM,并取对数。 |
| earnings_yield 盈利能力        | 定义：0.68 · EPFWD + 0.21 · CETOP + 0.11 · ETOP 其中 EPFWD: PE的预测值，用未来12个月的净利预测值除以当前市值。 CETOP: 过去12个月的现金分红除以当前的股票价格。 ETOP: 过去12个月的净利润除以当前的市值。 |
| growth 成长                    | 定义：0.18 · EGRLF + 0.11 · EGRSF + 0.24 · EGRO + 0.47 · SGRO 其中 EGRLF: 长期盈利增长预测，分析师的长期(未来3到5年) 净利润增长预测。 EGRSF: 短期盈利增长预测，分析师的长期(未来1年) 净利润增长预测。 EGRO: 过去5年净利润增长。 SGRO: 过去5年的销售额增长。 （注：由于暂时没有分析师预期数据，因此计算时将 EGRLF 与 EGRSF权重置为0） |
| leverage 杠杆                  | 定义： 0.38 · MLEV + 0.35o DTOA + 0.27 · BLEV 其中 MLEV: 市场杠杆，(当前的普通股市值+代先股账面价值+长期债务账面价值)/当前的普通股市值 DTOA: 资产负债比，总负债的账面价值/总资产的账面价值 BLEV: 账面杠杆，(普通股账面价値+优先股账面价值+长期债务账面价值)/普通股账面价值 |
